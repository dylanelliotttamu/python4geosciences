**What do all these input keyword arguments do?**

* header: tells which row of the data file is the header, from which it will extract column names


* parse_dates: try to interpret the values in `[col]` or `[col1, col2]` or `[co1, col2, col3, ...]` as dates, to convert the values of those columns into `datetime` objects.


* index_col: if no index column is given, an index counting from 0 to the (number of rows -1) is set up as the default index. By inputting `index_col=[column integer]`, that specific column will be used as the index instead. This is usually done for the column that containing the time information in the dataset.


* skiprows: can skip specific rows counting from the start (index 0 as the start) using `skiprows=[a list]`, or skip a number of rows from the start using `skiprows=N`. In our case, we don't skip any rows.


<div class="alert alert-info">
    
Note that <code>skiprows</code> will work first, and then <code>header = 0 </code> will count from the first row after your skipped rows!

Note that <code>skiprows = 0</code> is different than <code>skiprows = [0]</code>

## Data selection

One of the benefits of using `pandas` is being able to easily reference the data in intuitive ways.

### Select rows using iloc or loc

* Select data **by index number (i.e., the relative location)**, using `iloc`.
* Select data **by real index label (i.e., the values of the index itself)**, using `loc`.

```python
df
```

**iloc example**

```python
# select the first two rows
df.iloc[0:2,:]

# or
df.iloc[0:2]

# I suggest you stick to the first way as you will not struggle with whether 0:2 
# are the first two rows or the first two columns
```

```python
# select the first, the fifth and the tenth rows
df.iloc[[0, 4, 9], :]
```

**loc example**

```python
# select 2016-05-01 00:00:00 datetime
df.loc['2016-05-01 00:00:00', :]

# or
df.loc['2016-05-01 00:00:00', :]
```

can provide a date with specific components to `.loc`

```python
# select 2016-05-01 00 datetime (00 hour with all minutes and seconds)
df.loc['2016-05-01 00', :]
```

```python
df.loc['2016-05-01 00:05', :]
```

```python
df
```

```python
df.loc['2016-05-01 00:05':'2016-05-01 23:05', :]
```

**note that when using loc, the ending point is included (i.e., inclusive)**

```python
# select 2016-05-01 00 to 2016-05-01 05 datetime
df.loc['2016-05-01 00':'2016-05-01 05', :]
```

```python
# the same as above
df.loc[(df.index >= "2016-05-01 00") & (df.index <= "2016-05-01 05:59:59"), :]
```

# More plotting with pandas

```python
# ?df2.plot
```

You can mix and match plotting with matplotlib by two ways.

* either setting up a figure and axes in the matplotlib way, and then pass the axes to pandas `plot`.

* or you can fully use matplotlib in our normal way (highly recommended!)

## Start from `matplotlib`, then use axes in the pandas plot

You can select which data columns to plot either by selecting in the line before the `plot` call, or you can choose the columns within the plot call. 

The key part here is that you input to your pandas plot call the axes you wanted plotted into (here: `ax=axes[0]` or `ax=axs[1]`).

```python
df2.loc['2010-5', 'WSPD']
```

```python
fig, axs = plt.subplots(1, 2, figsize=(14,4))

# put ax inside plot()
df2.loc['2010-5', 'WSPD'].plot(ax=axs[0]) # note that this will not generate the legend by default
df2.loc['2010-5'].plot(y='WSPD', ax=axs[1]) # note that this will generate the legend by default
```

*How to add the legend in the first panel?*

```python
fig, axs = plt.subplots(1, 2, figsize=(14,4))

df2.loc['2010-5', 'WSPD'].plot(ax=axs[0], label="WSPD") # add label
axs[0].legend() # add legend

df2.loc['2010-5'].plot(y='WSPD', ax=axs[1])
```

How to delete the legend in the second panel?

```python
fig, axs = plt.subplots(1, 2, figsize=(14,4))

df2.loc['2010-5', 'WSPD'].plot(ax=axs[0])
df2.loc['2010-5'].plot(y='WSPD', ax=axs[1], legend=None) # get rid of legend
```

## Bring `pandas` dataframe data to `matplotlib` fully (highly recommended!)


```python
fig, ax = plt.subplots()

# the default axis tick labels can be messy
ax.plot(df2.loc['2010-5', 'WSPD'])
```

```python
fig, axs = plt.subplots(2, 2, figsize=(12,10))

# the default axis tick labels can be messy
axs[0,0].plot(df2.loc['2010-5', 'WSPD'])

# make the axis ticks pretty by limiting the number of tick labels
axs[0,1].plot(df2.loc['2010-5', 'WSPD'])
axs[0,1].xaxis.set_major_locator(plt.MaxNLocator(4))


# make the axis ticks pretty by rotating the labels
axs[1,0].plot(df2.loc['2010-5', 'WSPD'])
axs[1,0].tick_params(axis="x", rotation=30)

# make the axis ticks pretty by using the matplotlib.dates module
import matplotlib.dates as mdates
date_form = mdates.DateFormatter("%m %d")
ticklocations = mdates.DayLocator(interval=5)

axs[1,1].plot(df2.loc['2010-5', 'WSPD'])
axs[1,1].xaxis.set_major_formatter(date_form)
axs[1,1].xaxis.set_major_locator(ticklocations)
```

## Plot with twin y axis `ax.twinx()`

```python
fig, ax = plt.subplots(figsize=(14, 4))

ax.plot(df2.loc['2010-10', 'WSPD'], color="blue")

ax2 = ax.twinx()

ax2.plot(df2.loc['2010-10', 'WDIR'], color="orange")

ax.set_ylabel('Speed [m/s]', color='blue')
ax2.set_ylabel('Wind direction [degree]', color='orange')
```

```python
# add legend
# since we have two axes, we need to use fig.legend() to add the legend for the whole figure
fig, ax = plt.subplots(figsize=(14, 4))

ax.plot(df2.loc['2010-10', 'WSPD'], color="blue", label="wind speed")

ax2 = ax.twinx()

ax2.plot(df2.loc['2010-10', 'WDIR'], color="orange", alpha=0.5, label="wind direction")

ax.set_ylabel('Speed [m/s]', color='blue')
ax2.set_ylabel('Wind direction [degree]', color='orange')

fig.legend(loc="upper right", bbox_to_anchor=(0.8, 0.99), bbox_transform=ax.transAxes)
```

## Pandas data manipulation

### Resampling

Sometimes we want our data to be at a different sampling frequency that we have, that is, we want to change the time between rows or observations. Changing this is called resampling. We can upsample to increase the number of data points in a given dataset (or decrease the period between points) or we can downsample to decrease the number of data points.

The wind data is given every hour. Here we downsample it to be once a day instead. After the `resample` function, a method needs to be used for how to combine the data over the downsampling period since the existing data needs to be combined in some way. We could use the max value over the 1-day period to represent each day:

The codes for time period/frequency are [available](http://pandas.pydata.org/pandas-docs/stable/timeseries.html#offset-aliases) and are presented here for convenience:

    Alias	Description

    C	custom business day frequency (experimental)
    D	calendar day frequency
    W	weekly frequency
    M	month end frequency
    SM	semi-month end frequency (15th and end of month)
    BM	business month end frequency
    CBM	custom business month end frequency
    MS	month start frequency
    SMS	semi-month start frequency (1st and 15th)
    BMS	business month start frequency
    CBMS	custom business month start frequency
    Q	quarter end frequency
    BQ	business quarter endfrequency
    QS	quarter start frequency
    BQS	business quarter start frequency
    A	year end frequency
    BA	business year end frequency

### 2. upsampling

We can also upsample our data or add more rows of data. Note that like before, after we resample our data we still need a method on the end telling `pandas` how to process the data. However, since in this case we are not combining data (downsampling) but are adding more rows (upsampling), using a function like `max` doesn't change the existing observations (taking the max of a single row). For the new rows, we haven't said how to fill them so they are nan's by default.

Here we are changing from having data every hour to having it every 30 minutes.

```python
df2
```

```python
df2.resample('30min').max()  # NaN values will be generated
```

When upsampling, a reasonable option is to fill the new rows with data from the nearby row:

```python
# forward fill
df2.resample('30min').ffill()
```

```python
# backward fill
df2.resample("30min").bfill()
```

When upsampling, another reasonable option is to linearly interpolate:

```python
df2.resample('30min').interpolate()
```

### 3. rolling windows

Pandas `dataframe.rolling()` function provides the feature of rolling window calculations. The concept of rolling window calculation is most primarily used in signal processing and time series data. In a very simple words we take a window size of n at a time and perform some desired mathematical operation on it. A window of size n means n consecutive values at a time.

```python
df2
```

```python
df2.rolling(window=3).mean()
```

You can see that we get NaN in the first two rows because we cannot calculate the rolling mean as there are no preceding values to make the three-hour window complete. The third row value is the average value of each column over the previous three hours.

*what if we want the window to be centered around each point?*

```python
df2.rolling(window=3, center=True).mean()
```

We learned forward rolling and centered rolling, but how can we do backward rolling?

###  *Exercise*

> Do backward rolling with a window size of 3

```python
df2.iloc[::-1].rolling(window=3).mean().iloc[::-1]
```

```python
df2.rolling(window=3).mean().shift(-2)
```
