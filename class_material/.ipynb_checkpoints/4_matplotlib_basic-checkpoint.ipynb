## Figure and axes setup

Steps for setting up a figure:

1. Create a figure object using `plt.figure()`, specify the size of the figure, and save the figure object to a variable (here is **fig**)
2. Add axes to the figure object using `add_subplot()`, and save the axes object to a variable (here is **ax**). Note that Axes are the objects in which data is actually plotted.

        The complete format to add subplots is:

        fig.add_subplot(nrows, ncols, plot_number)
    
3. Plot! Most basically, use the `plot()` command to plot lines and markers.
4. Add labels to clearly explain the plot, such as axis labels and a title.

Let's make a plot:

```python
# create variables
x = np.arange(10, 20, 1)
y = x * 2

# create figure (based on figsize, width is 4, height is 4)
fig = plt.figure(figsize=(4,4))

# add axes (or a subplot)
# notes the 1,1,1 in the bracket
# first 1: the number of rows for our total subplots
# second 1: the number of columns for our total subplots
# third 1: which subplot we are plotting on now
ax = fig.add_subplot(1,1,1)

# plot
ax.plot(x, y)

# add label and title
ax.set_xlabel("x", fontsize=12)
ax.set_ylabel("y", fontsize=12)
ax.set_title('Title')
```

#### you can ignore the `1,1,1` inside add.subplot() if you only have 1 subplot.

```python
# create variables
x = np.arange(10, 20, 1)
y = x * 2

# create figure (based on figsize, width is 4, height is 4)
fig = plt.figure(figsize=(4,4))

# add axes (or a subplot)
ax = fig.add_subplot() # ignore 1, 1, 1

# plot
ax.plot(x, y)

# add label and title
ax.set_xlabel("x", fontsize=12)
ax.set_ylabel("y", fontsize=12)
ax.set_title('Title')
```

<div class="alert alert-danger">
Note that the above plotting method is called the <b>object-oriented approach</b>, and is highly recommended. There is a different way to plot using the <b>procedural approach</b> (which you may see a lot at some online tutorials), which I highly argue against!  </div>

**This is an example showing you the procedural approach. NOT RECOMMENDED!**

```python
# bad plotting method!!!
x = np.arange(10, 20, 1)
y = x * 2

plt.plot(x, y)
plt.xlabel("x", fontsize=12)
plt.ylabel("y", fontsize=12)
plt.title("y vs. x")
```

**Problems with the above procedural approach.**

* Less control
* Harder to make changes later
* Alters any figure you may already have open

<div class="alert alert-info">So, don't screw over future you! Set up your plot properly using the <b>object-oriented approach</b>!</div>

## Useful commands and keyword arguments

These commands and keyword arguments should be frequently used to customize and  label your figures. Command syntax shown is common usage, not all available options.

### More on labels and text

`ax.set_xlabel(xlabel, fontsize, color)`, `ax.set_ylabel(ylabel, fontsize, color)`: Label the x and y axis with strings xlabel and ylabel, respectively. This is where you should state what is being plotted, and also give units.

`ax.set_title(Title, fontsize, color)`: Label the top of the axes with a title describing the plot.

`ax.text(x, y, text, fontsize, color, transform=ax.transAxes)`: Write text in your axes. The text will appear at location (x,y) in data coordinates — often it is easier to input the location in units of the axes itself (from 0 to 1), which is done by setting `transform=ax.transAxes`. The text is input as a string and `color` controls the color of the text.

*let's try to add some text on the plot*

```python
x = np.arange(10, 20, 1)
y = x * 2

fig = plt.figure(figsize=(4,4))
ax = fig.add_subplot()

ax.plot(x, y)

ax.text(11, 36, "hallo", fontsize=20) # hard coded! not good.
```

```python
x = np.arange(10, 20, 1)
y = x * 2

fig = plt.figure(figsize=(4,4))
ax = fig.add_subplot()

ax.plot(x, y)

ax.text(0.1, 0.9, "hello", fontsize=20, transform=ax.transAxes) # relative position. good!
```

## Subplots

We have seen from above that we can add **one** subplot using `fig.add_subplot(1, 1, 1)` or `fig.add_subplot()`.

Recap: The complete format to add subplots is:

    fig.add_subplot(nrows, ncols, plot_number)

This `fig.add_subplot` command can be used to add multiple axes to the figure instead of a single one. 

An example with 1 row and 2 columns of axes is shown, with the `plot_number` increasing from 1 across the rows then columns, up to the total number of subplots or axes (2, in this case).

```python
# subplots example

x = np.arange(0, 10, 1)

len_x = len(x)
y1 = np.random.rand(len_x)
y2 = np.random.rand(len_x)

fig = plt.figure(figsize=(16, 4))

ax1 = fig.add_subplot(1, 2, 1)  # 1st subplot
ax1.plot(x, y1)
ax1.set_xlabel('x axis 1')
ax1.set_ylabel('y axis 1')
ax1.set_title('subplot 1', fontsize=14, color='orange')

ax2 = fig.add_subplot(1, 2, 2)  # 2nd subplot
ax2.plot(x, y2)
ax2.set_xlabel('x axis 2')
ax2.set_ylabel('y axis 2')
ax2.set_title('subplot 2', fontsize=14, color='orange')
```

### A better way to create multiple subplots:
fig, axs = plt.subplots(nrows, ncols, figsize, sharex, sharey) # for multiple subplots

<div class="alert alert-info">This function above allows us to 1) create fig and ax at the same time, 2) save all subplots into one single axes array. In addition, the subplots can have shared x, y, or both, by which the subplots then share the x and y limits and the ticks and tick labels.</div>

```python
# subplots example
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(16, 4), sharey=True)

axs[0].plot(x, y1)
axs[0].set_xlabel('x axis 1')
axs[0].set_ylabel('y axis 1')
axs[0].set_title('subplot 1', fontsize=14, color='orange')

axs[1].plot(x, y2)
axs[1].set_xlabel('x axis 2')
axs[1].set_title('subplot 2', fontsize=14, color='orange')
```

*How to add a total title for the whole figure?*
*use `fig.suptitle()`*

```python
# subplots example
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(16, 4), sharey=True)

axs[0].plot(x, y1)
axs[0].set_xlabel('x axis 1')
axs[0].set_ylabel('y axis 1')
axs[0].set_title('subplot 1', fontsize=14, color='orange')

axs[1].plot(x, y2)
axs[1].set_xlabel('x axis 2')
axs[1].set_title('subplot 2', fontsize=14, color='orange')

# use suptitle to add the total title
fig.suptitle('This is the overall title', fontsize=18, color="red")
```

*How to make the subplots tight and tidy?*
*use `fig.tight_layout()`, which is a convenience function to automatically improve spacing between subplots.*

```python
# subplots example
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(16, 4), sharey=True)

axs[0].plot(x, y1)
axs[0].set_xlabel('x axis 1')
axs[0].set_ylabel('y axis 1')
axs[0].set_title('subplot 1', fontsize=14, color='orange')

axs[1].plot(x, y2)
axs[1].set_xlabel('x axis 2')
axs[1].set_title('subplot 2', fontsize=14, color='orange')

# use suptitle to add the total title
fig.suptitle('This is the overall title', fontsize=18, color="red")

# use tight_layout to tidy the plots
fig.tight_layout()
```

*How to manullay adjust the margin of the subplots, and the padding between each subplot?*
*use `fig.subplots_adjust()`*

`fig.subplots_adjust(left=xx, bottom=xx, right=xx, top=xx, wspace=xx, hspace=xx)`: Any of the keywords listed may be used to override the default values. In order, they adjust the left, bottom, right, and top of the subplots, the width and height for space between subplots.

Note that left means "The position of the left edge of the subplots, as a fraction of the figure width.". And the same goes for bottom, right, and top.

```python
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(16, 4), sharey=True)

axs[0].plot(x, y1)
axs[0].set_xlabel('x axis 1')
axs[0].set_ylabel('y axis 1')
axs[0].set_title('subplot 1', fontsize=14, color='orange')

axs[1].plot(x, y2)
axs[1].set_xlabel('x axis 2')
axs[1].set_title('subplot 2', fontsize=14, color='orange')

# manually adjust the subplots
fig.subplots_adjust(left=0.1, bottom=0.1, right=0.8, top=0.8, wspace=0.2)
```

Let's try 6 subplots (with 3 rows and 2 columns). We demonstrate the ability to share both the x axis and y axis, and manually adjust the subplots.

## Plotting style

### Summary of the style options

- `color`: color of line
- `linestyle`: the line style
- `linewidth`: width of line if using a linestyle
<br><br>

- `marker`: the marker type
- `markersize`: size of the marker, if used
- `markerfacecolor`: color of marker face
- `markeredgecolor`: color of marker edge
- `markeredgewidth`: the line width of the marker edge.
<br><br>

- `alpha`: transparency of lines/markers, from 0 (transparent) to 1 (opaque)

### [colors](https://matplotlib.org/stable/api/colors_api.html)

A handful of colors are available by a single letter code:
- b: blue
- g: green
- r: red
- c: cyan
- m: magenta
- y: yellow
- k: black
- w: white, 
- and many more are available by (html) name (http://stackoverflow.com/questions/22408237/named-colors-in-matplotlib):


Other inputs to matplotlib possible:

* Gray scale: a `string` with a float in it between 0 (black) and 1 (white): for example `'0.3'` represents grey.
* RGB list in the range [0,1]: `[0.1, 0.2, 0.3]` (the original RGB space is ranging from 0 to 255 for each color, and we need to linearly scale them to 0-1 to be used in Python). The three numbers represent red, green and blue respectively.
* Hex color: `#EE99FF` (the first two digits represent red, the following two digits represent green, and the last two digits represent blue).

## Legends

A graph with multiple datasets may often benefit from a legend, which displays a label along each line/marker type somewhere within/along the figure.

A [legend](https://matplotlib.org/stable/api/legend_api.html) for a plot can be produced by matplotlib by labeling plots as they are plotted and then calling the legend command.

The basic command is:
    
    ax.legend(loc)
    
where loc tells where in the axes to place the legend:

    'best'         : 0,
    'upper right'  : 1,
    'upper left'   : 2,
    'lower left'   : 3,
    'lower right'  : 4,
    'right'        : 5,
    'center left'  : 6,
    'center right' : 7,
    'lower center' : 8,
    'upper center' : 9,
    'center'       : 10,

<div class="alert alert-info">You can simply add multiple plots on the same subplot!</div>

```python
x = np.arange(0, 10)

fig, ax = plt.subplots(figsize=(4,4))

# plot twice!
ax.plot(x, x)
ax.plot(x, x**2)
```

```python
x = np.arange(0, 10)

fig, ax = plt.subplots(figsize=(4,4))

ax.plot(x, x, label='y=x') # notice the label
ax.plot(x, x**2, label='y=x$^2$') # notice the label
ax.legend(loc='upper left')
```

*How can we make the labels in the legend in a single row?*
*use `ncol`*

```python
x = np.arange(0, 10)

fig, ax = plt.subplots(figsize=(4,4))

ax.plot(x, x, label='y=x') # notice the label
ax.plot(x, x**2, label='y=x$^2$',color='magenta') # notice the the label
ax.legend(loc='upper left', ncol=2, fontsize=15)
```

## ticks and ticklabels

Ticks are the locations of the little marks (ticklines) along the axis, and ticklabels are the text. You can control each of these independently in a couple of ways, including the following and with tick_params

```python
x = np.array([1,2,3,4,5,6,7]) # from Monday to Sunday

y = np.array([20, 18, 14, 20, 16, 20, 15]) # average temperature

fig, ax = plt.subplots(figsize=(8,4))

ax.plot(x, y)
```

#### change the fontsize of x ticklabel

```python
x = np.array([1,2,3,4,5,6,7]) # from Monday to Sunday

y = np.array([20, 18, 14, 20, 16, 20, 15]) # average temperature

fig, ax = plt.subplots(figsize=(8,4))

ax.plot(x, y)

# change fontsize of x ticklabels
ax.tick_params(axis="x", labelsize=20)
```

*How do we change the x ticklabels to Monday through Sunday?*

```python
x = np.array([1,2,3,4,5,6,7]) # from Monday to Sunday
y = np.array([20, 18, 14, 20, 16, 20, 15]) # average temperature of each day

fig, ax = plt.subplots(figsize=(8,4))

ax.plot(x, y)

# first, set the xticks
ax.set_xticks(ticks = [1, 2, 3, 4, 5, 6, 7]);

# second, customize the ticklabels corresponding to the ticks
ax.set_xticklabels(labels = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]);
```

Input a file name with file extension you want, fname, and the dots per inch (if saving to raster format, like png or jpeg) if you want something higher than the default. `bbox_inches='tight'` trims the white space around the figure; alternatively a number can be input to pad the figure if it cuts too close, but the 'tight' option works most of the time.

```python
x = np.array([1,2,3,4,5,6,7]) # from Monday to Sunday
y = np.array([20, 18, 14, 20, 16, 20, 15]) # average temperature of each day

fig, ax = plt.subplots(figsize=(8,4))

ax.plot(x, y)

# first, set the xticks
ax.set_xticks(ticks = [1, 2, 3, 4, 5, 6, 7]);

# second, customize the ticklabels corresponding to the ticks
ax.set_xticklabels(labels = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]);

fig.savefig("img_out/test_xx.pdf") # no dpi is needed

# save as pdf vector format 
# (note that your data_out folder needs to be in the same folder of your in-class jupyter notebook files)
fig.savefig("img_out/test.pdf", bbox_inches = 'tight') # no dpi is needed

# fig.savefig("img_out/test_xx.pdf") # no dpi is needed

# save as png raster format
# (note that your data_out folder needs to be in the same folder of your in-class jupyter notebook files)
fig.savefig("img_out/test.png", dpi=500, bbox_inches = 'tight') # assign a high dpi
```

### *Good design principles*

We all need to spend a bit of extra time to make really good plots instead of just-ok plots. This is worth the time because of how much clearer the result can be, which improves clarity, understanding, and communication.

[Edward Tufte](http://www.edwardtufte.com/tufte/) has books about how to think about design principles. A fundamental part of his thinking is to show the data without extra junk that doesn't add to the plot — he calls this extra plotting stuff "chartjunk". For example, if you have a plot with gridlines, make sure the gridlines are visible but not dominating the actual data.

Some guidelines to follow:
* Always label all axes, with units if appropriate;
* Make all text large enough to be easily seen by whatever type of viewer will be seeing your image — fontsize needs to be larger for a presentation than for a paper — this includes the ticklabels!;
* Make all lines thick enough to be easily seen;
* Make all lines/markers distinct in both color and line/marker style so that the legend is correct even in grayscale;
* Use colors that are complementary! The default choices aren't usually so pretty;
* Don't forget to pay attention to edge lines on markers and bars, the font style, and other details;


```python
x = np.linspace(0, 10)

fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))
axs[0].plot(x, x, label = 'x=x')
axs[0].plot(x, x**2, label = 'x=x$^2$')

axs[0].grid(True, color='r', linestyle='-', linewidth=2)
axs[0].set_title('Where is the data?\nHow would this look in black and white?')
axs[0].set_xlabel('xlabel [units]')
axs[0].set_ylabel('ylabel [units]')
axs[0].legend()

axs[1].plot(x, x, '0.3', linewidth=3, label = 'x=x')
axs[1].plot(x, x**2, '0.6', linewidth=4, linestyle='--', label = 'x=x$^2$')
axs[1].grid(True, color='k', linestyle=':', linewidth=0.25)
axs[1].set_title('Data displayed prominently and clearly', fontsize=14)
axs[1].set_xlabel('xlabel [units]', fontsize=14)
axs[1].set_ylabel('ylabel [units]', fontsize=14)
axs[1].legend(loc='best', fontsize=14)
```

### Labels with math

Sometimes we want to be able to use $\LaTeX$ to write math in axis labels, especially for units. We can do this in matplotlib! We just have to use proper $\LaTeX$ notation for writing the math, and put an 'r' in front of the string, though often the 'r' doesn't appear to be needed.

**Math notations or euqations (powerful!) (see tutorials of [this](https://www.codersarts.com/forum/machine-learning/writing-mathematical-equations-in-jupyter-notebook) and [this](https://math.meta.stackexchange.com/questions/5020/mathjax-basic-tutorial-and-quick-reference)):**

## Add colorbar with cmap

Matplotlib Colorbar is a visualization of the mapping of scalar values to colors. This way your data can be shown in a way to make it understandable to general audiences.

General steps:

1. assign `ax.plot(cmap=xx)` or `ax.scatter(cmap=xx)` to a variable (in our case it is `mappable`. You can name whatever you like)
2. create a `fig.colorbar()` object and assign it to a variable (in our case it is `cb`. You can name whatever you like)
3. set the properties of the colorbar.

Some good color maps for sequential data:

<img src="img/viridis_series.png" alt="viridis" width="50%"/>

#### horizontal colorbar

```python
x
```

```python
y
```

```python
x = np.random.uniform(size=50)
y = np.random.uniform(size=50)
colors = x + y**2 # create random colors based on x and y values


fig, ax = plt.subplots(figsize=(8,6))

# create the color map together with plotting the data
mappable = ax.scatter(x, y, marker="o", s=100, c=colors, cmap="viridis")

# set ax labels
ax.set_xlabel("x_value", fontsize=15)
ax.set_ylabel("y_value", fontsize=15)

# add colorbar (specify the mappable, ax, orientation and aspect)
cb = fig.colorbar(mappable, ax=ax, orientation="horizontal", pad=0.15, aspect=30)

# add the colorbar title
cb.set_label("colors", fontsize=20) # same as cb.ax.set_ylabel("colors", fontsize=20)

# change colorbar ticklabel size
cb.ax.tick_params(axis='x', labelsize=12, pad=10)
```

#### vertical colorbar

```python
fig, ax = plt.subplots(figsize=(8,6))

mappable = ax.scatter(x, y, marker="o", s=100, c=colors, cmap="viridis")

# set ax labels
ax.set_xlabel("x_value", fontsize=15)
ax.set_ylabel("y_value", fontsize=15)

cb = fig.colorbar(mappable, ax=ax, orientation="vertical", pad=0.05, aspect=30)

# add the colorbar title
cb.set_label("colors", fontsize=20) # same as cb.ax.set_ylabel("colors", fontsize=20)

# change colorbar ticklabel size
cb.ax.tick_params(axis='x',direction='in', labelsize=12, pad=10)
```

*plotting more than one subplot with two colorbars*

```python
# this is good!
fig, axs = plt.subplots(figsize=(16,6), nrows=1, ncols=2)

mappable1 = axs[0].scatter(x, y, c=colors, cmap="viridis", s=100, marker="o")

cb1 = fig.colorbar(mappable1, ax=axs[0], orientation="vertical", pad=0.05, aspect=30)

mappable2 = axs[1].scatter(x, y, c=colors, cmap="viridis", s=100, marker="o")

cb2 = fig.colorbar(mappable2, ax=axs[1], orientation="vertical", pad=0.05, aspect=30)
```

*However, it can get weird when plotting more than one subplot with a single colorbar*

```python
# if you only want one colorbar, the two subplots will have different widths!

fig, axs = plt.subplots(figsize=(16,6), nrows=1, ncols=2)

axs[0].scatter(x, y, c=colors, cmap="viridis", s=100, marker="o")

mappable2 = axs[1].scatter(x, y, c=colors, cmap="viridis", s=100, marker="o")

cb2 = fig.colorbar(mappable2, ax=axs[1], orientation="vertical", pad=0.05, aspect=30)
```

*How to fix?*
*: use the whole axs*

```python
fig, axs = plt.subplots(figsize=(16,6), nrows=1, ncols=2)

axs[0].scatter(x, y, c=colors, cmap="viridis", s=100, marker="o")

mappable2 = axs[1].scatter(x, y, c=colors, cmap="viridis", s=100, marker="o")

cb2 = fig.colorbar(mappable2, ax=axs, orientation="vertical", pad=0.05, aspect=30)
```

```python
# can also make horizontal one!
fig, axs = plt.subplots(figsize=(16,6), nrows=1, ncols=2)

axs[0].scatter(x, y, c=colors, cmap="viridis", s=100, marker="o")

mappable2 = axs[1].scatter(x, y, c=colors, cmap="viridis", s=100, marker="o")

cb2 = fig.colorbar(mappable2, ax=axs, orientation="horizontal", pad=0.1, aspect=60)
```
