## Spatial objects: Vector file vs. Raster file

There are two fundamental types of spatial objects: vectors and rasters. The main difference between raster and vector data is that the raster data represents data as a cell or a grid matrix while vector data represents data using sequential points or vertices. [Learn more](https://gisgeography.com/spatial-data-types-vector-raster/)

<img src="img/raster_vs_vector.png" width="500"/>


### Vector files

A commonly used vector data type is **shapefile**. Introduced in the early 1990s, the shapefile format is one of the most common GIS vector data formats compatible with the majority of software platforms. A shapefile record has a geometry, which contains the points/lines/polygons that make up the objects, and attributes, which store information like the name of the record. It typically contains information like the polygons describing counties or other political boundaries; lines describing rivers or coastlines, and points describing cities or lakes. More information about Shapefile can be found [here](https://www.gislounge.com/what-is-a-shapefile/).
 
<img src="img/spatial-attribute-tables.png" width="500"/>

### Data structure of shapefile
<img src="img/shapefile.png" width="500"/>

### Raster files

There are different raster data types, such as the popular NetCDF (.nc) and GeoTIFF (.tif).

1. NetCDF is a binary storage format for many different kinds of rectangular data. Examples include atmosphere and ocean model output, satellite images, and timeseries data. NetCDF files are intended to be device independent, and the dataset may be queried in a fast, random-access way. More information about NetCDF files can be found [here](https://www.unidata.ucar.edu/software/netcdf/).

2. GeoTIFF is a public domain metadata standard that enables georeferencing information to be embedded within an image file. The GeoTIFF format embeds geospatial metadata into image files such as aerial photography, satellite imagery, and digitized maps so that they can be used in GIS applications. More information about GeoTIFF files can be found [here](https://www.omnisci.com/technical-glossary/geotiff).

## Vector data analysis

We will maily use the **geopandas** package

* Geopandas is open-sourced library and enables the use and manipulation of geospatial data in Python. It extends the common datatype (`Series and DataFrame`) used in pandas to allow for the many and unique geometric operations: `GeoSeries and GeoDataFrame`.

* It combines the capabilities of **pandas** and **shapely**, providing geospatial operations in pandas and a high-level interface to multiple geometries to shapely. GeoPandas enables you to easily do operations in python that would otherwise require a spatial database such as PostGIS.

#### recommended method 1 for fine tuning: integrate geopandas plot with cartopy

steps:

1. create fig and ax using the cartopy way
2. plot using the geopandas way (passing in the cartopy ax; assigning the projection of the GeoDataFrame to the **transform parameter**)
3. modify the plot

```python
# 1. create fig and ax using the cartopy way
fig = plt.figure(figsize=(12, 12))

ax = fig.add_subplot(projection = ccrs.PlateCarree())

# 2. plot using the geopandas way
gdf_land.plot(ax=ax, transform=ccrs.PlateCarree(), facecolor="grey", zorder=3) # zorder to make the gridlines behind the land

# modify the plot
gl = ax.gridlines(draw_labels=True, 
                  linewidth=0.5,
                  x_inline=False, 
                  y_inline=False, 
                  xformatter=LONGITUDE_FORMATTER, 
                  yformatter=LATITUDE_FORMATTER)

gl.xlabel_style = {'size': 15, 'color': 'gray'}  # control how the tick labels look
gl.ylabel_style = {'size': 15, 'color': 'gray'}
gl.top_labels = False  # turn off labels where you don't want them
gl.right_labels = False
```

#### recommend method 2 for fine tuning: use pure cartopy methods

Steps:

1. create fig and ax using the cartopy way
2. use `ax.add_geometries()` to add the geometry of the GeoDataFrame (assigning the projection of the GeoDataFrame to the **crs parameter**)
3. modify the plot


Notes:
1. We need to use the `gdf_land['geometry']` inside `ax.add_geometries()`, not `gdf_land` itself). 
2. We need to use `crs=the_projection_of_your_geometry`, not `transform=the_projection_of_your_geometry` inside `ax.add_geometries()`!

```python
fig = plt.figure(figsize=(12, 12))

ax = fig.add_subplot(projection = ccrs.PlateCarree())

ax.add_geometries(gdf_land['geometry'], crs = ccrs.PlateCarree(), facecolor="grey", zorder=3)

gl = ax.gridlines(draw_labels=True, 
                  linewidth=0.5,
                  x_inline=False, 
                  y_inline=False, 
                  xformatter=LONGITUDE_FORMATTER, 
                  yformatter=LATITUDE_FORMATTER)

gl.xlabel_style = {'size': 15, 'color': 'gray'}  # control how the tick labels look
gl.ylabel_style = {'size': 15, 'color': 'gray'}
gl.top_labels = False  # turn off labels where you don't want them
gl.right_labels = False
```

#### we need to use an alternative way to create the colorbar (using `norm`)

no legend:

```python
fig = plt.figure(figsize=(12, 12))

ax = fig.add_subplot(projection = ccrs.PlateCarree())

gdf_country.plot(ax=ax, transform = ccrs.PlateCarree(), column='POP_EST', cmap='viridis', zorder=3)

gl = ax.gridlines(draw_labels=True, 
                  linewidth=0.5,
                  x_inline=False, 
                  y_inline=False, 
                  xformatter=LONGITUDE_FORMATTER, 
                  yformatter=LATITUDE_FORMATTER)

gl.xlabel_style = {'size': 15, 'color': 'gray'}  # control how the tick labels look
gl.ylabel_style = {'size': 15, 'color': 'gray'}
gl.top_labels = False  # turn off labels where you don't want them
gl.right_labels = False
```

let's add the legend:

```python
fig = plt.figure(figsize=(12, 12))

ax = fig.add_subplot(projection = ccrs.PlateCarree())

gdf_country.plot(ax=ax, transform = ccrs.PlateCarree(), column='POP_EST', cmap='viridis', zorder=3)

gl = ax.gridlines(draw_labels=True, 
                  linewidth=0.5,
                  x_inline=False, 
                  y_inline=False, 
                  xformatter=LONGITUDE_FORMATTER, 
                  yformatter=LATITUDE_FORMATTER)

gl.xlabel_style = {'size': 15, 'color': 'gray'}  # control how the tick labels look
gl.ylabel_style = {'size': 15, 'color': 'gray'}
gl.top_labels = False  # turn off labels where you don't want them
gl.right_labels = False

################# add the legend

# create norm of pop for color
norm = plt.Normalize(vmin=min(gdf_country.POP_EST), vmax=max(gdf_country.POP_EST))

# manually add colorbar
mappable = plt.cm.ScalarMappable(norm=norm, cmap='viridis')

cb = fig.colorbar(mappable, ax=ax, orientation="horizontal", pad=0.1, shrink=1, aspect=30)

# add the colorbar title
cb.set_label("populations", fontsize=20, labelpad=-70)

cb.ax.tick_params(axis='x',direction='in', labelsize=14, length=5, width=1)
```

<div class="alert alert-info">
    
To sum up, when plotting geopandas objects, you can choose either `gdf.plot(ax=ax)` or `ax.add_geometries(gdf['geometry'])`. When colorbar is involved, `gdf.plot(ax=ax)` is recommended!

#### different projections

```python
gdf_country.crs
```

One of the most powerful websites to look up CRS strings is [Spatialreference.org](https://spatialreference.org/). You can use the search on the site to find an EPSG code. Once you find the page associated with your CRS of interest you can then look at all of the various formats associated with that CRS: [EPSG 4326 - WGS84 geographic](https://spatialreference.org/ref/epsg/4326/).

```python
# create a projection (equal area)
ea = ccrs.LambertAzimuthalEqualArea()
ea.proj4_init # this is called proj4 string, a very useful crs format!!!
```

```python
# change the crs of gdf_country to this equal area proj4 string
# and assign the result to a new variable
gdf_country_lambert = gdf_country.to_crs(ea.proj4_init)
```

```python
gdf_country.crs
```

```python
gdf_country_lambert.crs
```

```python
# integrate cartopy with geopandas plot
fig = plt.figure(figsize=(12, 12))

ax = fig.add_subplot(projection = ccrs.LambertAzimuthalEqualArea())

gdf_country_lambert.plot(ax=ax, transform=ccrs.LambertAzimuthalEqualArea(), column='POP_EST', cmap='viridis', zorder=3) # zorder to make the gridlines behind the land

gl = ax.gridlines(draw_labels=True, 
                  linewidth=0.5,
                  x_inline=False, 
                  y_inline=False, 
                  xformatter=LONGITUDE_FORMATTER, 
                  yformatter=LATITUDE_FORMATTER)

gl.xlabel_style = {'size': 15, 'color': 'gray'}  # control how the tick labels look
gl.ylabel_style = {'size': 15, 'color': 'gray'}
gl.top_labels = False  # turn off labels where you don't want them
gl.right_labels = False

################# add the legend

# create norm of pop for color
norm = plt.Normalize(vmin=min(gdf_country.POP_EST), vmax=max(gdf_country.POP_EST))

mappable = plt.cm.ScalarMappable(cmap='viridis', norm=norm)

cb = fig.colorbar(mappable, ax=ax, orientation="horizontal", pad=0.1, shrink=1, aspect=30)

# add the colorbar title
cb.set_label("populations", fontsize=20, labelpad=-70)

cb.ax.tick_params(axis='x',direction='in', labelsize=14, length=5, width=1)
```

#### DataFrame to GeoDataFrame

In our daily study, our data might not be in the shapefile format. We need to convert from DataFrame object to GeoDataFrame object

```python
# get our sedimentation rate
df_sed = pd.read_csv('data/sedimentation_rate_part.csv')
```

```python
df_sed.head()
```

```python
type(df_sed)
```

```python
# convert from dataframe to geodataframe
# use gpd.points_from_xy to convert the lon and lat to geometry
gdf_sed = gpd.GeoDataFrame(df_sed, 
                           geometry=gpd.points_from_xy(df_sed.lon, df_sed.lat), 
                           crs=4326)
```

```python
gdf_sed.head()
```

```python
type(gdf_sed)
```

```python
gdf_sed.crs
```

```python
# now we plot

fig = plt.figure(figsize=(12, 12))

ax = fig.add_subplot(projection = ccrs.PlateCarree())

ax.coastlines(linewidth=0.5)

# plot sedimentation rate
gdf_sed.plot(ax=ax, transform=ccrs.PlateCarree(), column='sedimentation_rate', cmap='viridis', zorder=3, s=10, edgecolor="white", linewidth=0.1)

gl = ax.gridlines(draw_labels=True, 
                  linewidth=0.5,
                  x_inline=False, 
                  y_inline=False, 
                  xformatter=LONGITUDE_FORMATTER, 
                  yformatter=LATITUDE_FORMATTER)

gl.xlabel_style = {'size': 15, 'color': 'gray'}  # control how the tick labels look
gl.ylabel_style = {'size': 15, 'color': 'gray'}
gl.top_labels = False  # turn off labels where you don't want them
gl.right_labels = False

# add colorbar

# create the norm of sedimentation rate for color
norm_sed = plt.Normalize(vmin=min(gdf_sed.sedimentation_rate), vmax=max(gdf_sed.sedimentation_rate))

mappable = plt.cm.ScalarMappable(cmap='viridis', norm=norm_sed)

cb = fig.colorbar(mappable, ax=ax, orientation="horizontal", pad=0.1, shrink=1, aspect=30)

# add the colorbar title
cb.set_label("Sedimentation rate (mm/yr)", fontsize=20, labelpad=-70)

cb.ax.tick_params(axis='x',direction='in', labelsize=14, length=5, width=1)
```

```python
# Hurricane Katrina path

# a list of latitudes and a list of longitudes (lons, lats) for Hurricane Katrina (2005)

lons = [-75.1, -75.7, -76.2, -76.5, -76.9, -77.7, -78.4, -79.0,
            -79.6, -80.1, -80.3, -81.3, -82.0, -82.6, -83.3, -84.0,
            -84.7, -85.3, -85.9, -86.7, -87.7, -88.6, -89.2, -89.6,
            -89.6, -89.6, -89.6, -89.6, -89.1, -88.6, -88.0, -87.0,
            -85.3, -82.9]

lats = [23.1, 23.4, 23.8, 24.5, 25.4, 26.0, 26.1, 26.2, 26.2, 26.0,
            25.9, 25.4, 25.1, 24.9, 24.6, 24.4, 24.4, 24.5, 24.8, 25.2,
            25.7, 26.3, 27.2, 28.2, 29.3, 29.5, 30.2, 31.1, 32.6, 34.1,
            35.6, 37.0, 38.6, 40.1]
```

```python
# we create a line shapefile
# if you still remember how we convert scattered points to points geometry , we use
# geometry=gpd.points_from_xy(df_sed.lon, df_sed.lat). Unfortunately, we don't have 
# such method to convert scattered points to line geometry.
# Instead, we need to use the shapely LineString and the zip function
gdf_katrina =  gpd.GeoDataFrame(geometry = [LineString(coordinates=zip(lons, lats))], 
                                crs=4326)
```

```python
gdf_katrina
```

```python
gdf_katrina.plot()
```

```python
# now we read the U.S. states
gdf_states = gpd.read_file("data/ne_110m_admin_1_states_provinces")
```

```python
gdf_states.head()
```

make sure gdf_states and gdf_katrina are in the same projection:

```python
gdf_states.crs
```

```python
gdf_katrina.crs
```

```python
gdf_states.plot()
```

```python
# plot both states and the katrina path
fig = plt.figure(figsize=(12, 12))
ax = fig.add_subplot(projection = ccrs.PlateCarree())

gdf_states.plot(ax=ax, transform = ccrs.PlateCarree())

gdf_katrina.plot(ax=ax, transform = ccrs.PlateCarree(), color="red")
```

**now we do the intersection**

```python
gdf_katrina.geometry
```

```python
gdf_katrina
```

```python
# get the polygon
# since gdf_katrina only has 1 row of data (the katrina line path),
# .iloc[0] will extract the whole path
gdf_katrina.geometry.iloc[0]
```

```python
# obtain the boolean index that intersects with katrina
# note that we need to use gdf_katrina.geometry[0] to avoid warning
intersect_index = gdf_states.intersects(gdf_katrina.geometry.iloc[0])
```

```python
type(intersect_index)
```

```python
intersect_index
```

```python
# use boolean index to filter the states
gdf_states_intersect = gdf_states.loc[intersect_index, ]
```

```python
gdf_states_intersect.name
```

```python
# now we plot

fig = plt.figure(figsize=(12, 12))
ax = fig.add_subplot(projection = ccrs.PlateCarree())

ax.coastlines(linewidth=0.5)

gdf_states.plot(ax=ax, transform = ccrs.PlateCarree(), facecolor="grey", edgecolor="white", linewidth=0.2)

# add intersected states
gdf_states_intersect.plot(ax=ax, facecolor="blue", edgecolor="white", linewidth=0.2)

# add katrina line
gdf_katrina.plot(ax=ax, color="red", linewidth=2, facecolor="none")


gl = ax.gridlines(draw_labels=True, 
                  linewidth=0.5,
                  x_inline=False, 
                  y_inline=False, 
                  xformatter=LONGITUDE_FORMATTER, 
                  yformatter=LATITUDE_FORMATTER)

gl.xlabel_style = {'size': 15, 'color': 'gray'}  # control how the tick labels look
gl.ylabel_style = {'size': 15, 'color': 'gray'}
gl.top_labels = False  # turn off labels where you don't want them
gl.right_labels = False

# set limits to only show us region
ax.set_extent([-130, -60, 20, 55], crs = ccrs.PlateCarree())
```

```python
# buffer the linestring by two degrees (note: this is a non-physical distance and is only for demonstration purpose)
# normally, we will buffer using real distances (in km or m)
gs_katrina_buffer = gdf_katrina.buffer(2) # 2 is for degrees here
```

```python
gs_katrina_buffer.geometry.iloc[0]
```

```python
# intersect states with the buffer
intersect_buffer_index = gdf_states.intersects(gs_katrina_buffer.geometry.iloc[0])
```

```python
# filter the states that intersect with the buffer
gdf_states_intersect_buffer = gdf_states.loc[intersect_buffer_index, ]
```

```python
# now we plot

fig = plt.figure(figsize=(12, 12))
ax = fig.add_subplot(projection = ccrs.PlateCarree())

ax.coastlines(linewidth=0.5)

gdf_states.plot(ax=ax, facecolor="grey", edgecolor="white", linewidth=0.2)

# add intersected buffered states
gdf_states_intersect_buffer.plot(ax=ax, transform = ccrs.PlateCarree(), facecolor="orange", edgecolor="white", linewidth=0.2)


# add intersected states
gdf_states_intersect.plot(ax=ax, transform = ccrs.PlateCarree(), facecolor="blue", edgecolor="white", linewidth=0.2)

# add buffered katrina
gs_katrina_buffer.plot(ax=ax, transform = ccrs.PlateCarree(), facecolor="orange", alpha=0.3, linewidth=0.2)

# add katrina
gdf_katrina.plot(ax=ax, transform = ccrs.PlateCarree(), color="red", linewidth=2, facecolor="none")


gl = ax.gridlines(draw_labels=True, 
                  linewidth=0.5,
                  x_inline=False, 
                  y_inline=False, 
                  xformatter=LONGITUDE_FORMATTER, 
                  yformatter=LATITUDE_FORMATTER)

gl.xlabel_style = {'size': 15, 'color': 'gray'}  # control how the tick labels look
gl.ylabel_style = {'size': 15, 'color': 'gray'}
gl.top_labels = False  # turn off labels where you don't want them
gl.right_labels = False

# set limits
ax.set_extent([-130, -60, 20, 55], crs = ccrs.PlateCarree())
```
