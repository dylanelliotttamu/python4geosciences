```python
X, Y = np.meshgrid(np.linspace(-4.2, 6.2, 101), np.linspace(-3.2, 3.2, 101))
Z = np.cos(X**2+Y**2)*(X**2+Y**2)**-0.25 + 0.2*X*Y*np.sin(Y) \
   + 0.01*X*Y**4*np.sin(X) + 0.2*np.exp(-Y**2)

fig, axs = plt.subplots(nrows=1, ncols=3,figsize=(15,4))
# axs.set_title('sfg')
map0 = axs[0].pcolormesh(X,Y,Z, cmap='plasma')
map1 = axs[1].contourf(X,Y,Z, cmap='plasma')
map2 = axs[2].contour(X,Y,Z, cmap='plasma')
axs[0].set_aspect(1)
axs[1].set_aspect(1)
axs[2].set_aspect(1)
axs[2].clabel(map2)

'''colorbars'''

# mappable1 = axs[1].pcolormesh(Z, cmap='viridis')
cb0 = fig.colorbar(map0, ax=axs[0], orientation="horizontal")
cb1 = fig.colorbar(map1, ax=axs[1], orientation="horizontal")
cb2 = fig.colorbar(map2, ax=axs[2], orientation="horizontal")

tick_font_size = 12

cb0.set_label("Z", fontsize=tick_font_size)
cb1.set_label("Z", fontsize=tick_font_size)
cb2.set_label("Z", fontsize=tick_font_size)

cb0.ax.tick_params(labelsize=tick_font_size)
cb1.ax.tick_params(labelsize=tick_font_size)
cb2.ax.tick_params(labelsize=tick_font_size)

```

### 2. 2D interpolation (6 points)

Data (found in `MS03_L15.txt`) gathered from a field cruise out of the TAMU Oceanography department is randomly located in space (that is, unstructured) because of the nature of the measurement: the instrument is lowered into the water from a ship, and moves up and down through the water column as the ship moves to get a sampling of the fields.

1. Read in the data file (you can use either `np.loadtxt(xxx, ...)` or `pd.read_table(xxx, ...)`. Then plot the salinity in a `scatter` plot as a function of depth and latitude, with salinity as color. Be sure that the colors in the markers are visible.

1. Next, interpolate the salinity data to be on a uniform grid of depth and latitude (to make it structured). Hint: when you use meshgrid on depth and latitude, try to use np.linspace to generate a sequence of depth and latitude first (see our class material for reference). If you directly use meshgrid on the original non-structured depth and latitude, meshgrid might crash you laptop due to huge memory consumption. When you use griddata(), think carefully which method you want to use: cubic, linear, or nearest (certain methods might yield unrealistic salinity values on the edge of your plots)

1. Now, plot the salinity using `contourf`, which requires structured data. Experiment with different spacing in your uniform grid, and with different numbers of contours to see how the results change, and choose values that provide a nice-looking image which closely resembles the original data.

All plots should be in a reasonable orientation given what the data represents in reality. Make sure your axis label and legend fontsize is big enough.

Below is what I made for your reference. Note that you should have your own style.

<img src="p2.png" width=600 />

---

Grading rubric [6 points]:

* [2 pts] Scatter plot: correct, complete, and nice-looking
* [2 pts] Interpolation for gridding
* [2 pts] contourf plot: correct, complete, and nice-looking

```python
ocean_data = pd.read_table('MS03_L15.txt', comment='*', header=None, delim_whitespace=True)

ocean_data.columns = (['Lat','Long','Depth','Temp','density00','sigma-t00','Oxygen','v2','v3','v4','wetCDOM','flECO','upoly0','Pressure','sal00','RINKO TEMP','Rinko do','rinkodo2','bottom depth','PAR'])

''' salinity in a scatter plot as a function of depth and latitude, with salinity as color '''

fig, axs = plt.subplots(nrows=1,ncols=2,figsize=(18,6),sharex=True,sharey=True)
axs[0].invert_yaxis()
salinity = ocean_data['sal00']
mappable = axs[0].scatter(ocean_data['Lat'],ocean_data['Depth'], marker="o", s=50, c=salinity, cmap="viridis")
axs[0].set_ylabel('Depth',fontsize=16)
axs[0].set_xlabel('Latitude',fontsize=16)

'''interpolate'''
lat = ocean_data['Lat']
dep = ocean_data['Depth']

lat_i = np.linspace(lat.min(),lat.max(),100)
dep_i = np.linspace(dep.min(),dep.max(),100)

L , D = np.meshgrid(lat_i, dep_i)
sal_grid = griddata((lat,dep),salinity,(L,D), method = 'linear', fill_value=np.nan)

sal_mesh = axs[1].contourf(L, D, sal_grid, cmap='viridis')
axs[1].set_xlabel('Latitude',fontsize=16)

cb = fig.colorbar(sal_mesh,ax=axs)
cb.set_label('Salinity',fontsize=16)

axs[0].set_title('Raw Data\n',fontsize=20)
axs[1].set_title('Interpolated Data\n',fontsize=20)
```
